```python
reference_sentences = [
    "The cat sits on the mat.",
    "A quick brown fox jumps over the lazy dog.",
    "Thanksgiving is a holiday with a lot of food.",
]

hypothesis_sentences = [
    "A cat is resting on the rug.",
    "The quick brown fox leaps over the sleeping dog.",
    "Thanksgiving is a holiday with no food.",
]
```

```python
idf_dict_ref = get_idf_dict(reference_sentences)
idf_dict_hyp = get_idf_dict(hypothesis_sentences)
```

```python
scores = word_mover_score(
    reference_sentences, hypothesis_sentences, idf_dict_ref, idf_dict_hyp
)
print("Word Mover's Scores:", scores)
```
