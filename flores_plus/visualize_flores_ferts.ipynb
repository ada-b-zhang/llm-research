```python
import pandas as pd
import matplotlib.pyplot as plt

flores_ferts = pd.read_csv('../../flores/flores_fertilities.csv')
flores_ferts.head()
```

```python
flores_ferts.drop(columns=['id']).describe()
```

```python
fig, ax = plt.subplots()
boxprops = dict(color="red", facecolor="pink")
boxplot = ax.boxplot([flores_ferts['aya101'], 
                      flores_ferts['bloom560'],
                      flores_ferts['flan_t5'],
                      flores_ferts['llama3'],
                      flores_ferts['mt0']], 
                      labels=['aya101', 'bloom560', 'flan_t5', 'llama3', 'mt0'], 
                      showfliers=False, patch_artist=True, showmeans=True,
                      meanprops={'marker': '*',               
                                 'markerfacecolor': 'green',
                                 'markeredgecolor': 'green', 
                                 'markersize': 10})

for box in boxplot['boxes']:
    box.set(facecolor="gold", edgecolor="green")  

for median in boxplot['medians']:
    median.set(color="green", linewidth=2)  

plt.suptitle('Fertility Comparison of Different Tokenizers')
plt.title('(without outliers)')
plt.xlabel('Model/Tokenizer')
plt.ylabel('Fertility Score')
plt.show()
```
