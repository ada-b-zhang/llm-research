```python
import pandas as pd
import matplotlib.pyplot as plt

flores_ferts = pd.read_csv('flores_fertilities.csv').drop(columns=['Unnamed: 0'])
flores_ferts.head()
```

```python
flores_ferts.drop(columns=['id']).describe()
```

```python
fig, ax = plt.subplots(figsize=(12,8))
boxplot = ax.boxplot([flores_ferts['Aya-101'], 
                      flores_ferts['Bloomz-7B1'],
                      flores_ferts['FlanT5-XXL'],
                      flores_ferts['Llama-3.2-1B-Instruct'],
                      flores_ferts['mT0-XXL-MT'],
                      flores_ferts['Phi-3.5-mini-instruct'],
                      flores_ferts['Pangea-7B'],
                      flores_ferts['Gemma-7B']], 
                      labels=['Aya-101', 'Bloomz-7B1', 'FlanT5-XXL', 'Llama-3.2-1B-Instruct', 'mT0-XXL-MT', 
                              'Phi-3.5-mini-instruct', 'Pangea-7B', 'Gemma-7B'], 
                      showfliers=False, patch_artist=True, showmeans=True,
                      meanprops={'marker': '*',               
                                 'markerfacecolor': 'green',
                                 'markeredgecolor': 'green', 
                                 'markersize': 10})

for box in boxplot['boxes']:
    box.set(facecolor="gold", edgecolor="green")  

for median in boxplot['medians']:
    median.set(color="green", linewidth=2)  

ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

plt.title('Fertility Comparison of Different Tokenizers (WITHOUT outliers)')
plt.xlabel('Model/Tokenizer')
plt.ylabel('Fertility Score')
plt.show()
```
